```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
```

```python
# Configure headless Chrome for Colab
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(options=chrome_options)

driver.get("https://ayoubcomputers.com/laptops/")
time.sleep(3)
```

```python
brands = []
titles = []
prices = []


while True:
    time.sleep(2)
    # Get all laptop products (each product is an <li> element with class "product")
    products = driver.find_elements(By.CSS_SELECTOR, 'li.product')

    for product in products:
        # Extract Brand
        try:
            brand = product.find_element(By.CSS_SELECTOR, 'p.card-text.card-text--brand').text.strip()
        except Exception as e:
            brand = ""

        # Extract Title
        try:
            title = product.find_element(By.CSS_SELECTOR, 'h4.card-title a').text.strip()
        except Exception as e:
            title = ""

        # Extract Price
        try:
            price = product.find_element(By.CSS_SELECTOR, 'span.price--main').text.strip()
        except Exception as e:
            price = "not available"

        # Append data in the order: Brand, Title, Price
        brands.append(brand)
        titles.append(title)
        prices.append(price)

        print(f"Brand: {brand}\nTitle: {title}\nPrice: {price}\n{'-'*40}")

    # ---  Handle Pagination ---
    try:
        next_button = driver.find_element(By.XPATH, '//li[contains(@class, "pagination-item--next")]/a')
        next_button.click()
    except Exception as e:
        print("✅ Reached last page.")
        break

driver.quit()
```

```python
# --- Save Data to Excel ---
df = pd.DataFrame({
    'Title': titles,
    'Brand': brands,
    'Price': prices
})
df.to_excel("laptops.xlsx", index=False)
print("✅ Scraping complete! File saved as laptops.xlsx")
```

```python
# Filter for laptops with no price
no_price_laptops = [i for i in zip(titles, brands, prices) if i[2] == ""]

# Print laptops without price
for laptop in no_price_laptops:
    print(f"Title: {laptop[0]}\nBrand: {laptop[1]}\nPrice: {laptop[2]}\n{'-'*40}")
```
